```python
import pandas as pd
import matplotlib.pyplot as plt

log_file_path ='log.csv'
df = pd.read_csv(log_file_path, names=["Name", "Month", "Year", "Department", "Income"])

avg_income_by_dept = df.groupby("Department")["Income"].mean()

plt.figure(figsize=(10, 6))
avg_income_by_dept.plot(kind="bar")
plt.title("Average Income by Department")
plt.xlabel("Department")
plt.ylabel("Average Income")

plt.show()
```
